# Uczenie Nadzorowane

Uczenie nadzorowane (ang. supervised learning) to jeden z głównych rodzajów uczenia maszynowego, w którym model jest szkolony na podstawie danych, które zawierają zarówno wejście (cechy) jak i pożądane wyjście (etykiety lub docelowe wartości). Celem uczenia nadzorowanego jest nauczenie modelu rozpoznawania lub przewidywania odpowiednich wyjść na podstawie nowych, wcześniej niewidzianych danych. Jest to proces polegający na nauczaniu modelu na podstawie przykładów, aby model mógł generalizować i dokonywać predykcji na nowych danych.

#### Główne cechy uczenia nadzorowanego to:

1. **Dane treningowe**: W uczeniu nadzorowanym potrzebujemy zestawu danych treningowych, który składa się z par cechy-wyjście. Cecha to opisująca dane informacja, podczas gdy wyjście to oczekiwana odpowiedź lub etykieta. Te dane treningowe są używane do nauki modelu.

2. **Model**: Model to matematyczna reprezentacja, która przekształca cechy wejściowe na wyjścia. Model jest inicjowany z pewnymi parametrami, które są dostosowywane podczas procesu uczenia w celu minimalizacji błędu predykcji.

3. **Funkcja straty**: Funkcja straty (ang. loss function) jest używana do oceny, jak dobrze model dokonuje predykcji w porównaniu do rzeczywistych etykiet w danych treningowych. Celem jest minimalizacja funkcji straty, co oznacza, że model stara się być jak najbliżej rzeczywistych wyjść.

4. **Algorytm uczący**: Algorytm uczący to procedura lub metoda, która dostosowuje parametry modelu w oparciu o dane treningowe i funkcję straty. Popularnym algorytmem w uczeniu nadzorowanym jest metoda gradientu prostego (gradient descent), która minimalizuje funkcję straty poprzez dostosowywanie parametrów modelu w kierunku malejącego gradientu funkcji straty.

5. **Proces uczenia**: Proces uczenia polega na wielokrotnym dostosowywaniu parametrów modelu na podstawie danych treningowych. Model jest trenowany tak długo, aż funkcja straty osiągnie odpowiednio niską wartość lub do momentu, gdy model osiąga satysfakcjonującą wydajność na danych treningowych.

6. **Walidacja i testowanie**: Po treningu modelu konieczne jest ocenienie jego wydajności na nowych danych, które nie były używane w procesie uczenia. W tym celu dane są podzielone na zbiór treningowy, walidacyjny i testowy. Zbiór walidacyjny jest używany do dostosowywania hiperparametrów modelu, podczas gdy zbiór testowy jest używany do oceny jego ogólnej wydajności.

7. **Generalizacja**: Celem uczenia nadzorowanego jest osiągnięcie ogólnej zdolności modelu do dokonywania trafnych predykcji na nowych, nieznanych danych. Proces ten nazywany jest generalizacją, a model jest oceniany pod kątem jego zdolności do radzenia sobie z nowymi danymi.

Uczenie nadzorowane znajduje zastosowanie w wielu dziedzinach, takich jak klasyfikacja, regresja, rozpoznawanie wzorców, przetwarzanie języka naturalnego i wiele innych. Jest to jeden z najpopularniejszych rodzajów uczenia maszynowego ze względu na swoją zdolność do rozwiązywania różnorodnych problemów, w tym problemów klasyfikacji, prognozowania, detekcji, i wiele innych.

## Klasyfikacja

#### Klasyfikacja binarna
> Model na podstawie danych ma zakwalifikować wynik do jednej z dwóch kategorii np. SPAM / NIE SPAM

#### Klasyfikacja wieloklasowa
> Model na podstawie danych ma zakwalifikować wynik do wielu zdefiniowanych kategorii np. Pozytywne / Neutralne / Negatywne  
> Można zdefiniować prawdopodobieństwo przypisania dla każdej z kategorii  
> Można wytrenować model (neuronowe sieci konwolucyjne) który będzie klasyfikował te same dane do kategorii różnych typów np. Typ: sukienka, Kolor: czerwony




## Regresja

Regresja to przewidywanie wartości ciągłej zmiennej na podstawie posiadanych danych np. przewidywanie cen mieszkań na podstawie danych o mieszkaniach

## Workflow uczenia maszynowego

1. Określenie problemu biznesowego
2. Pozyskanie danych
3. Analiza danych
4. Przygotowanie danych
5. Dobór odpowiednich modeli
6. Strojenie modeli i wybór najbardziej optymalnego
7. Wdrożenie modelu i ciągły monitoring
